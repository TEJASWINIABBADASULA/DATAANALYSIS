```python
df3=df2.copy()
df3=df3.dropna()
print(df3)
```

```python
df3.dropna(inplace=True,axis=1)#inplace says not to be reflect the change in the df
print(df3)
```

```python
df3.dropna(inplace=True,how='all')#drop both rows and columns with NaN
print(df3)
```

```python
mv = df3['SEM1'].mean()
df3=df3.fillna(mv)
print(df3)
```

```python
df3 = df3.drop_duplicates()
print(df3)
```

```python
df3.rename(columns={'Grade':'GPA'},inplace=True)
df3.head()
```

```python
df3.columns = ['Student_name', 'Mr/Ms', '1stsem', '2ndsem', '3rdsem', 'F Outcome', 'Placed?']
df3.head()
```

```python
df3['Avg score'] = (df3['1stsem']+df3['2ndsem']+df3['3rdsem'])/3
df3.head()
```
